### Import all the required libraires.

```python
import numpy as np    #Library to handle data in vectorized manner.
import pandas as pd   #library for data analysis.

#Plotting libray matplotlib and associated ploting modules.
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as colors

#import folium geo map library.
import folium

%matplotlib inline
print('All the required libraries are imported.....!')
```

### For Predicted result dataset :

#### Read the data from csv in pandas dataframe.


```python
result = pd.read_csv("Result.csv")
result.head()
```

#### Remove the blank cells.

```python
nan_value = float("NaN")
result.replace("", nan_value, inplace=True)
result.dropna(subset = ["District_name"], inplace=True)
print("Drop Sucessfull")
#now look at the shape of our data frame.
print("Size of the data frame is :", result.shape)
```

#### Set District name as index.

```python
#Set the district name as index - useful for quickly looking up district using .loc method
result.set_index('District_name' , inplace = True)

#now see how data frame looks.
result.head()
```

#### Now Transpose the dataframe :

```python
T_result = result.T

T_result.head()
```

#### Rename the index names.

```python
#rename index.
T_result.rename(index ={ 'Predicted_rice': 'RICE', 'Predicted_wheat': 'WHEAT', 'Predicted_jowarr' : 'JOWARR', 'Predicted_bajra': 'BAJRA', 'Predicted_maize': 'MAIZE', 'Predicted_ragi': 'RAGI', 'Predicted_gram': 'GRAM', 'Predicted_tur': 'TUR', 'Predicted_udid': 'UDID', 'Predicted_mung': 'MUNG'}, inplace=True)
T_result.head()
```

### Now Select the relavent data from prediction result dataset as per user input.

Please Enter the Crop name in which you want to analyze crop data in the input field, use the Crop name from following list.
[RICE, WHEAT , JOWARR, BAJRA, MAIZE, RAGI, GRAM, TUR, UDID, MUNG]

NOTE : Input field is case sensitive.

```python
value = input("Enter the Crop Name :")
value
```

#### Get and preprocess data.

```python
df_pred = T_result.loc[[value]] #assign the data of district which is selected by user input to the new dataframe.

#Transpose Dataframe.
df_pred = df_pred.T

#make index as column.
df_pred = df_pred.reset_index()

df_pred.head()
```

### Lets view the Geographic map of Maharashtra, India with predicted Crop yield.

```python
maha_geo = r'maha.geojson' # geojson file

# create a plain world map
maha_map = folium.Map(location=[19.663280, 75.300293], zoom_start=7, tiles='Mapbox Bright')
```

```python
# generate choropleth map using the Crop yield Prediction in the maharashtra state.
maha_map.choropleth(
    geo_data=maha_geo,
    data=df_pred,
    columns=['District_name', value],
    key_on='feature.properties.name_2',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Predicted Crop Yield in Maharashtra State (Values in Tonns).'
    
)

# display map
maha_map
```

#### Save Map.

```python
#Save map.
maha_map.save("Map.html")
```
